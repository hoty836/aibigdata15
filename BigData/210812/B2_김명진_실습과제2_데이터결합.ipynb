# 데이터 결합 (p.32)

## 1. 데이터 구성하기

```python
import pandas as pd

# df4, df5, df6 데이터 생성
df4 = pd.DataFrame({
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]},
        index=[0, 1, 2, 3])

df5 = pd.DataFrame({
        "E": ["E4", "E5", "E6", "E7"],
        "F": ["F4", "F5", "F6", "F7"],
        "G": ["G4", "G5", "G6", "G7"],
        "H": ["H4", "H5", "H6", "H7"]},
        index=[0, 1, 2, 3])

df6 = pd.DataFrame({
        "I": ["I8", "I9", "I10", "I11"],
        "J": ["J8", "J9", "J10", "J11"],
        "K": ["K8", "K9", "K10", "K11"],
        "L": ["L8", "L9", "L10", "L11"]},
        index=[0, 1, 2, 3])
```

```python
display(df4, df5, df6)
```

## 2. concat을 이용한 데이터 결합

```python
result = pd.concat([df4, df5, df6])
result
```

## 3. concat의 parameter 변경

```python
# axis = 1 : 가로 결합
result1 = pd.concat([df4, df5, df6], axis = 1)
result1
```

```python
result2 = pd.concat([df4, df5, df6], axis = 1, ignore_index = True)
#result2.column = ['col_' + str(i) for i in range(12)]
for i in range(12):
    result2.rename(columns = {result2.columns[i]:'col_'+str(i)}, inplace = True)
result2
```

- 교재에 표기된 것처럼 주석처리된 코드를 통해 column명을 변경하려고 하였으나 userwarning이 뜨면서 새로운 속성 이름을 통해 열들이 생성되도록 pandas가 허용하지 않는다는 메시지가 떴다. 
- 그래서 찾아본 결과 rename 메소드를 알게 되었고 이를 이용하여 ignore_index = True를 통해 새로운 index로 재생성하여 concat 시키고 rename을 통해 column명을 변경하여보았다.
