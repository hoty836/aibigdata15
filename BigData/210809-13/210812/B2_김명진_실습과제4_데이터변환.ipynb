# 데이터 변환 (p.54)
- pivot을 이용하여 GENDER와 AGEGROUP별 WEIGHT의 평균을 교재와 같이 변환하라.

## 1. 데이터 구성하기

```python
import pandas as pd

fit=pd.read_csv("C:/Users/Myung-jin/Git_Temp-main/FITNESS.CSV")
fit
```

## 1-1 데이터 가공하기

```python
# GENDER 그룹의 2번째 행에 값 채우기
fit.iloc[1, 1] = 'F'
fit.head()
```

```python
# 잘못 설정된 AGEGROUP 수정해주기
fit.iloc[:, 3] = fit.iloc[:, 2] - (fit.iloc[:,2]%10)
fit
```

```python
# 특정열만 불러오기
fit1 = fit[['GENDER', 'AGEGROUP', 'WEIGHT']]
fit1
```

## 2. groupby, agg를 이용한 평균값 계산

```python
df_group = fit1.groupby(['GENDER', 'AGEGROUP']).agg('mean')
df_group
```

```python
df_group.reset_index(inplace = True)
df_group
```

```python
# 데이터값 수정하기
df_group.replace('F', '여성', inplace = True)
df_group.replace('M', '남성', inplace = True)
df_group
```

```python
df_group.replace({'AGEGROUP' : {30: '30대'}}, inplace = True)
df_group.replace({'AGEGROUP' : {40: '40대'}}, inplace = True)
df_group.replace({'AGEGROUP' : {50: '50대'}}, inplace = True)
df_group
```

```python
df_group.iloc[:, 2] = df_group.iloc[:, 2].round(3)
df_group
```

## 3. pivot을 이용한 변환

```python
# pivot을 이용하여 index는 GENDER, column은 AGEGROUP으로 설정하여 변환
df_group = df_group.pivot(index = 'GENDER', columns = 'AGEGROUP', values = 'WEIGHT')
df_group
```

원래 데이터에서 해당하는 수치의 mean값을 구해보았더니 81.030이 맞았다. 교재와 수치가 완전하게 일치하진 않지만 문제의 원인을 해결할 수 없다고 결론을 냈고, 데이터 수치나 교재 자료의 오류라고 판단한다.**
