## 1. 데이터 구성하기
- 두 데이터를 merge를 이용하여 FITNESS 데이터로 만들어라.

```python
import pandas as pd
fit1=pd.read_csv("C:/Users/Myung-jin/Git_Temp-main/FITNESS_1.csv",encoding='cp949')
fit2=pd.read_csv("C:/Users/Myung-jin/Git_Temp-main/FITNESS_2.csv")

display(fit1, fit2)
```

## 2. merge를 이용한 결합

```python
result1 = pd.merge(fit1, fit2, on=['NAME', 'AGE', 'GENDER'])
result1.head()
```

## 3. 원본 FITNESS 데이터와 비교

```python
fit=pd.read_csv("C:/Users/Myung-jin/Git_Temp-main/FITNESS.csv", encoding='cp949')
fit.head()
```

- AGEGROUP의 ~대를 제외해야 원본과 같아진다.

## 4. 원본 FITNESS 데이터와 같게 만들기

```python
result1.dtypes
```

```python
fit.dtypes
```

- 데이터 타입을 확인해본 결과 result1의 AGEGROUP은 '~대'이기 때문에 obj, 원본 데이터의 AGEGROUP은 타입이 int64이다. 따라서 result1의 AGEGROUP의 value를 수정하여 저장하고자 한다.

```python
result1.AGEGROUP = result1.AGEGROUP.str.replace('대', '')
```

```python
result1.head()
```

```python
result1.dtypes
```

- 눈으로 볼 때는 원본데이터와 똑같아졌지만, data type을 봤을 때 여전히 object이다. 형태를 int64로 변환해본다.

```python
import numpy as np
result1.AGEGROUP = result1.AGEGROUP.astype(np.int64)
result1.dtypes
```

**두 데이터를 merge와 가공을 통해 result1을 생성하였고, 원본 데이터 FITNESS와 동일하게 만들었다.**
